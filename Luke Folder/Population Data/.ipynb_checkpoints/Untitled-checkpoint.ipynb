```python
csv_path = Path("../Resources/Houston-population-2023-07-27.csv")
pd.read_csv(csv_path)
pop_hou=pd.read_csv(csv_path)
pop_hou.head()
```

```python
#Remove unwanted information
rows_to_remove = list(range(0, 58)) + list(range(74, 86))
pop_hou_dropped = pop_hou.drop(index=rows_to_remove)
pop_hou_dropped
```

```python
print(pop_hou_dropped.columns)
```

```python
x_axis = np.arange(len(pop_hou_dropped))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(10,4))
plt.bar(x_axis, pop_hou_dropped[" Population"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, pop_hou_dropped["Date"], rotation="vertical")

plt.xlabel('Date')
plt.ylabel('Population')
plt.title('Population Over Time')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(pop_hou_dropped['Date'], pop_hou_dropped[' Population'], marker='o', linestyle='-', color='b')

plt.xlabel('Date')
plt.ylabel('Population')
plt.title('Population Over Time')
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```
