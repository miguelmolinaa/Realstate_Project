```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
csv_files = [
    "../Resources/2017-2021 American Community Survey 5-Year Estimates.csv",
    "../Resources/2012-2016 American Community Survey 5-Year Estimates.csv"
]

csv_path = Path("../Resources/2017-2021 American Community Survey 5-Year Estimates.csv")
pd.read_csv(csv_path)
pop_df21=pd.read_csv(csv_path)
pop_df21.head()
 
```

```python
pop_df21.columns
```

```python
new_column_names = {
    'Label (Grouping)': 'Age',
    'Houston-The Woodlands-Sugar Land, TX Metro Area!!Total!!Estimate': 'Houston-The Woodlands-Sugar Land, TX Metro Area!!Total!!Estimate',
    'Houston-The Woodlands-Sugar Land, TX Metro Area!!Percent!!Estimate': 'Houston-The Woodlands-Sugar Land, TX Metro Area!!Percent!!Estimate',
    'Houston-The Woodlands-Sugar Land, TX Metro Area!!Male!!Estimate': 'Male Estimate',
    'Houston-The Woodlands-Sugar Land, TX Metro Area!!Percent Male!!Estimate': 'Male Percent Estimate',
    'Houston-The Woodlands-Sugar Land, TX Metro Area!!Female!!Estimate': 'Female Estimate',
    'Houston-The Woodlands-Sugar Land, TX Metro Area!!Percent Female!!Estimate': 'Female Percent Estimate'
}

pop_df21.rename(columns=new_column_names, inplace=True)
```

```python
pop_df21.head()
```

```python
columns_to_hide = ['Houston-The Woodlands-Sugar Land, TX Metro Area!!Total!!Estimate', 'Houston-The Woodlands-Sugar Land, TX Metro Area!!Percent!!Estimate']
df_filtered = pop_df21.drop(columns=columns_to_hide)

df_filtered.head()
```

```python
selected_rows = df_filtered.iloc[2:20]
selected_rows
```

```python
selected_columns = ['Age', 'Male Percent Estimate']
df_selected = selected_rows[selected_columns].copy()

# Replace '(X)' values with NaN using .loc
df_selected.loc[df_selected['Male Percent Estimate'] == '(X)', 'Male Percent Estimate'] = np.nan

# Remove percentage symbol (%) and convert to numeric using .loc
df_selected['Male Percent Estimate'] = df_selected['Male Percent Estimate'].str.rstrip('%')
df_selected['Male Percent Estimate'] = df_selected['Male Percent Estimate'].astype(float)

# Set 'Age' column as the index for easier plotting
df_selected.set_index('Age', inplace=True)

# Create the bar chart with the desired figure size
#plt.figure(figsize=(9, 8))
df_selected.plot(kind='bar', color='skyblue', legend=None)
plt.title('Male Percent Estimate Across Age Groups 2017-2021')
plt.xlabel('Ages')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
selected_columns = ['Age', 'Female Percent Estimate']
df_selected = selected_rows[selected_columns].copy()

# Replace '(X)' values with NaN using .loc
df_selected.loc[df_selected['Female Percent Estimate'] == '(X)', 'Female Percent Estimate'] = np.nan

# Remove percentage symbol (%) and convert to numeric using .loc
df_selected['Female Percent Estimate'] = df_selected['Female Percent Estimate'].str.rstrip('%')
df_selected['Female Percent Estimate'] = df_selected['Female Percent Estimate'].astype(float)

# Set 'Age' column as the index for easier plotting
df_selected.set_index('Age', inplace=True)

# Create the bar chart with the desired figure size
#plt.figure(figsize=(9, 8))
df_selected.plot(kind='bar', color='pink', legend=None)
plt.title('Male Percent Estimate Across Age Groups 2017-2021')
plt.xlabel('Ages')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```
