```python
min_latitude = df_crime['MapLatitude'].min()
max_latitude = df_crime['MapLatitude'].max()
min_longitude = df_crime['MapLongitude'].min()
max_longitude = df_crime['MapLongitude'].max()

# Display the calculated boundaries
print("Minimum Latitude:", min_latitude)
print("Maximum Latitude:", max_latitude)
print("Minimum Longitude:", min_longitude)
print("Maximum Longitude:", max_longitude)
```

```python
# Calculate the latitude and longitude range for each box
latitude_range = (max_latitude - min_latitude) / 6
longitude_range = (max_longitude - min_longitude) / 6

# Create a list to store the bounding boxes for each box (16 in total)
boxes = []
for i in range(6):
    for j in range(6):
        # Calculate the boundaries for each box
        min_lat_box = min_latitude + (i * latitude_range)
        max_lat_box = min_latitude + ((i + 1) * latitude_range)
        min_lon_box = min_longitude + (j * longitude_range)
        max_lon_box = min_longitude + ((j + 1) * longitude_range)

        # Append the boundaries to the boxes list
        boxes.append((min_lat_box, max_lat_box, min_lon_box, max_lon_box))

# Display the bounding boxes for each box
for idx, box in enumerate(boxes, 1):
    print(f"Box {idx}: Latitude Range: {box[0]} to {box[1]}, Longitude Range: {box[2]} to {box[3]}")
```

```python
# Create a new column 'Box' in the DataFrame to indicate which box each crime belongs to
def assign_box(latitude, longitude):
    for idx, box in enumerate(boxes, 1):
        if box[0] <= latitude <= box[1] and box[2] <= longitude <= box[3]:
            return f"Box {idx}"
    return "Unknown"

df_crime['Box'] = df_crime.apply(lambda row: assign_box(row['MapLatitude'], row['MapLongitude']), axis=1)

# Count the number of crimes in each box
crime_counts = df_crime.groupby('Box').size()

# Display the crime counts for each box
print(crime_counts)
```

```python
import pandas as pd
import folium

# file path of your CSV file for df_crime
file_path2 = 'Recourses_miguel/NIBRSPublicViewJun23.csv'

# Read the CSV file into a pandas DataFrame
df_crime = pd.read_csv(file_path2)

# Drop the specified columns
columns_to_drop = ['Incident', 'RMSOccurrenceDate', 'RMSOccurrenceHour', 'NIBRSClass', 'OffenseCount', 'Beat', 'Premise', 'Suffix']
df_crime = df_crime.drop(columns=columns_to_drop)

# Filter out rows with missing (NaN) values in either 'MapLatitude' or 'MapLongitude' columns
df_crime = df_crime.dropna(subset=['MapLatitude', 'MapLongitude'])

# Create a folium map centered at a specific location (you can choose the center based on your data)
map_center = [29.749907, -95.358421]  # Replace latitude_center and longitude_center with the desired center point
crime_map = folium.Map(location=map_center, zoom_start=10)
counter = 0
# Plot latitude and longitude points on the map as small dots (CircleMarkers) (limit to 100 iterations)
for index, row in df_crime.iterrows():
    latitude = row['MapLatitude']
    longitude = row['MapLongitude']
    folium.CircleMarker([latitude, longitude], radius=2, fill=True, color='blue', fill_opacity=0.6).add_to(crime_map)
    counter += 1
    if counter >= 5000:
        break

# Display the map
crime_map
```
